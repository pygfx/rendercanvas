# RenderCanvas notebook examples

## Cube example

```python
# from rendercanvas.anywidget import RenderCanvas
# from rendercanvas.jupyter import RenderCanvas
from rendercanvas.auto import RenderCanvas  # Defaults to anywidget when in a notebook
from rendercanvas.utils.cube import setup_drawing_sync

canvas = RenderCanvas(update_mode="continuous")
draw_frame = setup_drawing_sync(canvas)
canvas.request_draw(draw_frame)

canvas
```

```python
# Set title to non-empty string to show the title bar
canvas.set_title("Rotating cube")
canvas.set_minimizable(True)
canvas.set_closable(True)
```

## Snapshots

It is also possible to create snapshots. This is also possible without displaying the interactive canvas itself. The benefit of snapshots is that the output remains visible when the notebook is shown in a static viewer. This makes them useful for e.g. tutorials.

```python
canvas.snapshot()
```

## Events

```python
import ipywidgets

out = ipywidgets.Textarea(rows=10)


@canvas.add_event_handler("*")
def show_events(event):
    if event["event_type"] not in ["pointer_move", "before_draw", "draw"]:
        out.value = str(event)


out
```
